```
files = glob.glob("chart_*.csv")
result = []
for f in files:    

    df = pd.read_csv(f)
    if len(df.index)==8800:
        top_2 = pd.DataFrame(columns = ['Region','Top Artists','Artist Count','Top Songs','Song Count',])
        n_of_wks = np.ceil(len(df.index)/200)

        sa = df['artist'].value_counts().nlargest(c)
        ss = df['title'].value_counts().nlargest(c)

        df_artist = pd.DataFrame({'artist':sa.index, 'count':sa.values})
        df_song = pd.DataFrame({'song':ss.index, 'count':ss.values})
 

        top_2['Top Artists'] = df_artist['artist']
        top_2['Top Songs'] = df_song['song']
        top_2['Artist Count'] = df_artist['count']
        top_2['Song Count'] = df_song['count']
        top_2['Region'] = f.split('_')[1]
        top_2 = top_2.merge(df, left_on = 'Top Songs', right_on = 'title', how = 'left')
        top_2 = top_2.groupby(['Region', 'Top Artists', 'Artist Count', 'Top Songs', 'Song Count']).mean(
            ['rank', 'streams', 'acousticness', 'danceability', 'energy', 'loudness','speechiness', 'tempo', 'valence']
                    ).reset_index()

        result.append(top_2)
    
```
